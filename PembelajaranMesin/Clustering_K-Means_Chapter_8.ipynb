```python
import numpy as np
import pandas as pd
```

```python
nama_file = "data_kebun_binatang.xlsx"
```

```python
path_dir = ".\\"
DataFrame = pd.read_excel( path_dir + nama_file, encoding = "cp1252", sep = ';' )
df = DataFrame.copy(deep=True)
df.head(5)
```

```python
column_yang_tidak_dipakai_pada_file = 'nama binatang'

if column_yang_tidak_dipakai_pada_file in df.columns:
    df = df.drop(column_yang_tidak_dipakai_pada_file,1)
df.head(5)
```

```python
from sklearn.cluster import KMeans

jumlah_clustering = 7
kmeans = KMeans(n_clusters=jumlah_clustering, random_state=1)
data_kmeans_baru =  df._get_numeric_data()
kmeans.fit(data_kmeans_baru)
prediksi = kmeans.predict(data_kmeans_baru)
df_kmeans = df.copy(deep=True)
df_kmeans['Clusters KMeans'] = pd.Series(prediksi, index = df_kmeans.index)
df_kmeans.head(20)
```

```python
#Kemudian kita simpan ke file data kebun binatang
df_kmeans_original = DataFrame.copy(deep=True)
df_kmeans_original['Clusters KMeans'] = pd.Series(prediksi,index = df_kmeans.index)

#Kita akan memindahkan columnt cluster ke depan 
colom = df_kmeans_original.columns.tolist()
colom = colom[0:1] + [colom[-1]] + colom[1:-2]
df_kmeans_original = df_kmeans_original[colom]

#tambahkan ke dataframe
df_kmeans_original = df_kmeans_original.fillna("")

file_kmeans = nama_file[0:4] + "_kmeans_" + str(jumlah_clustering) + ".csv"
path_kmeans = path_dir +file_kmeans
df_kmeans_original.to_csv(path_or_buf = path_kmeans, sep=";")
print("Data sudah tersimpan")
```
