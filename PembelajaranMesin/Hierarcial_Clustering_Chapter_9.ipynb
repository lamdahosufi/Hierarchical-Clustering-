```python
from sklearn.cluster import AgglomerativeClustering
```

```python
from sklearn.datasets import make_blobs
```

```python
X,y = make_blobs(random_state=0, n_samples=1000)
```

```python
X
```

```python
import matplotlib.pyplot as plt
```

```python
plt.plot(X[:,0],X[:,1],'ro')
plt.show()
```

```python
clustering = AgglomerativeClustering(n_clusters=3, linkage="ward", affinity='euclidean')
```

```python
clustering.fit(X)
```

```python
plt.scatter(X[:,0],X[:,1], c=clustering.labels_, marker='o')
plt.show()
```

# Postest_9

Complete - Manhattan

```python
clustercomplete = AgglomerativeClustering(n_clusters=3, linkage="complete", affinity='manhattan')
```

```python
clustercomplete.fit(X)
```

```python
plt.scatter(X[:,0],X[:,1], c=clustercomplete.labels_, marker='o')
plt.show()
```

Average - Manhattan

```python
clusteraverage = AgglomerativeClustering(n_clusters=3, linkage="average", affinity='manhattan')
```

```python
clusteraverage.fit(X)
```

```python
plt.scatter(X[:,0],X[:,1], c=clusteraverage.labels_, marker='o')
plt.show()
```
